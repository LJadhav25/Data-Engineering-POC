# **Usecase-2** Display only those customers from CSV_s1, who purchased more than 3 products

```python
new = pd.merge(customer,transaction)
new_group = new.groupby(['custno'])['custno'].count()
three_products = new_group[new_group > 3].index
filtered_data = customer[customer['custno'].isin(three_products)]
filtered_data
```

# **Usecase-3** Display top 5 most demanded products from CSV_s1

```python
pro_trans = pd.merge(product,transaction)
product_counts = pro_trans['productno'].value_counts()
top_5_products = product_counts.head(5)
out = pd.merge(product,top_5_products,how='inner',left_on=['productno'],right_on=['productno'])
out
```

# **Usecase-4** Display top 5 transactions amount from CSV_s1

```python
trans = transaction.groupby(['custno'])['amount'].sum()
five_largest = trans.nlargest(5)
last_output = pd.merge(transaction,five_largest,how='inner',left_on=['custno'],right_on=['custno'])
last_output.drop_duplicates(subset=['custno'])
```

# **Usecase-7** Display custno, gender, age, profession, contactno, productno, productName, txndate, spendby, amount from CSV_s1 for custno = 923

```python
get_cust = customer[customer['custno'] == 923]
get_trans = pd.merge(get_cust,transaction)
semi_correct = get_trans[['custno','gender','age','profession','contactNo','productno','txndate','spendby','amount']]
final_cut = pd.merge(semi_correct,product)
final_cut[['custno','gender','age','profession','contactNo','productno','productName','txndate','spendby','amount']]
```
